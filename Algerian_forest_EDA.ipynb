```python
fig, axes = plt.subplots(2,2, figsize=(12,10) )
si1 = [[0,0], [0,1], [1,0], [1,1]]

for item in range(len(weatherlist)): 
    i = si1[item][0]
    j = si1[item][1]
    sns.lineplot(data = data, x = 'month', y = weatherlist[item], hue = 'Classes1', style = 'Region', ax =axes[i,j])
    axes[i,j].set_title(f'Month vs {weatherlist[item]}')
    axes[i,j].set_xlabel('Month')
    axes[i,j].set_ylabel(weatherlist[item])
    
    
plt.show() 
```

## Observations :
1. Temperature increases from the month of june and reaches to peak in august and then decreases by 
   september. Also, when temperature is in the range from 31 to > 34, more chances of fire incident. 
2. Temperature and relative humidity are negatively correlated. Hence, less the humdity more are the chances 
   for the fire incident.
3. For positive fire incidents Windspeed decreases with month. Windspeed is lesser for positive fire incidents 
than no fire incidents. In the month of august if 15<Ws<16, lesser probability of fire incidents and if Ws < 15
more the probability of fire incidents.  
4. Fire incidents are more probable with lesser rain. Rain is lesser in july than june. Then it increases by the
month of august and remains kind of constant throught august september.
